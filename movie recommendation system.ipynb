```python
import pandas as pd
import numpy as np
```

Merge Dataframes on the Title from different files

```python
movies = pd.read_csv('J:\\My Drive\\Coding\\Data Science Project\\Movie Recomendation\\dataset\\tmdb_5000_movies.csv')

credits = pd.read_csv('dataset\\tmdb_5000_credits.csv')

movies= movies.merge(credits, on = 'title')
movies.sample(5)
```

Feature Selection

```python
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.head()
```

Data Extraction: The genres and keywords are stored as "Stringified Lists of Dictionaries." You need a helper function to convert these into clean lists of strings.

```python
import ast
def convert(text):

    L = []

    for i in ast.literal_eval(text):
        L.append(i['name'])
    return L

movies['genres'] = movies['genres'].apply(convert)
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.dropna(inplace= True)
```

```python
movies.head(5)
```

```python
def convert3(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L 
```

Targeted Info Extraction

```python
def fetch_director(text):

    L = []

    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])

    return L

movies['crew'] = movies['crew'].apply(fetch_director)
movies['cast'] = movies['cast'].apply(lambda x: x[0:3])
```

```python
movies
```

Entity Resolution (Removing Spaces)

```python
def collapse(L):
    L1 = []

    for i in L:
        L1.append(i.replace(" ", ""))
    return L1

movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
```

Step 6: Feature Fusion (Creating the Tags)

We now combine all our processed attributes into a single pool of words called tags. We also split the overview string into a list of words so it can be concatenated with our other lists. Finally, we join the big list back into one long string.

```python
movies['overview'] = movies['overview'].apply(lambda x: x.split())
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']


#   new dataframe with just ID, Title, and the fused Tags
new = movies.drop(columns=['overview', 'genres', 'keywords', 'cast', 'crew'])
new['tags'] = new['tags'].apply(lambda x: " ".join(x))
```

Step 7: Text Vectorization (The Bag of Words)

We need to convert these paragraphs into a mathematical format. We use the Bag of Words model. It creates a matrix where each row is a movie and each column is one of the 5,000 most frequent words. The cells contain the count of that word in that movie.

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words = 'english')
vector = cv.fit_transform(new['tags']).toarray()
```

Step 8: Cosine Similarity (Calculating Distance)

In a 5,000-dimensional space, we want to see how close two movies are. We use Cosine Similarity, which measures the angle between vectors. An angle of 0° (Similarity = 1) means they are identical; 90° (Similarity = 0) means they are completely unrelated.

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vector)
```

Step 9: The Recommendation Function - The final logic:

1.  Find the index of the requested movie.

2.  Get the similarity scores for that index.

3.  Use enumerate to keep the movie indices attached to their scores.

4.  Sort by score and pick the top 5 (excluding the movie itself).

```python
def recommend(movie):
    index = new[new['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])), reverse= True, key= lambda x: x[1])

    for i in distances[1:6]:
        print(new.iloc[i[0]].title)

recommend('Gandhi')
```
